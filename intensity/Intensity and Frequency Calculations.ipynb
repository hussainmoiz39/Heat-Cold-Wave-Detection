Heat Wave Intensity & Frequency Calulation

```python
for i in range(0,len(lat)):
    x = lat[i]
    y = lon[i]
    print 'Latitude: ',x, 'Longitude: ', y
    df1 = list(df.loc[ (df['Lat']==x) & (df['Lon']==y) & (df['EHF'] > 0) ]['EHF'] )
    df2 = list(df.loc[ (df['Lat']==x) & (df['Lon']==y) ]['EHF'] )
    df1.sort()
    #print len(df1)
    #print df1[-25:]
    mean = sum(df1)/len(df1)
    #print mean 
    idx = float( (0.75)*len(df1) )
    if (idx).is_integer():
        t10 = (df1[ int(idx)-1] + df1[int(idx)])/2
    else:
        t10 = df1[int(idx)-1]

    #print idx, len(df1), t10,mean
    intensity = []
    cnt = 0
    #print len(df1)

    for i in range(0, len(df2)):
        if df2[i] >= t10:
            cnt = 1 + cnt
        else:
            if cnt > 0:
                intensity.append(cnt)
                cnt = 0
            else:
                cnt = 0

    print 'Frequency: ', len(intensity)  
    print 'Duration of Intensities: ', intensity
    print " "

    
    
    
```
